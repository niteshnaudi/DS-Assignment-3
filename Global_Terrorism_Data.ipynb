# **3. Data Wrangling**

## Data Wrangling Code

```python
import pandas as pd

# Load dataset
df = pd.read_csv("Global Terrorism Data.csv", encoding='latin1')

# Check basic info
print(df.shape)
print(df.info())

# Check missing values
print(df.isnull().sum())

# Drop columns with more than 50% missing values
df = df.dropna(axis=1, thresh=len(df)*0.5)

# Fill remaining missing values
df.fillna("Unknown", inplace=True)

# Check duplicates
print(df.duplicated().sum())

# Check unique values
print(df.nunique())
```

## What all manipulations have you done and insights you found?

=> Data Manipulations:Loaded the dataset using pandas
Checked dataset structure using info() and shape
Identified missing values using isnull()
Removed columns with more than 50% missing values
Filled remaining null values with "Unknown"
Checked for duplicate rows (none found)
Analyzed unique values in each column

### Chart-15

```python
import numpy as np

labels = ['Kills', 'Wounds', 'Success']
values = [
    df['nkill'].mean(),
    df['nwound'].mean(),
    df['success'].mean()
]

angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)

values = np.concatenate((values, [values[0]]))
angles = np.concatenate((angles, [angles[0]]))

plt.polar(angles, values)
plt.fill(angles, values, alpha=0.3)

plt.xticks(angles[:-1], labels)
plt.title("Radar Chart of Attack Features")
plt.show()
```

### 1. Why did you pick this chart?

A radar chart is used to compare multiple features (kills, wounds, success) in one view, showing their relative magnitude.

### 2. Insights from the chart:
Compares average kills, wounds, and success rate.
Helps identify which factor is highest or most dominant.
Shows overall pattern of attack severity and effectiveness.
### 3. Business Impact:

Positive:

Helps in understanding key impact factors of attacks.
Useful for planning security and emergency response.

Negative:

High values indicate serious threat levels.
Can lead to economic loss, reduced investment, and instability.
